#EXPERIMENT-3_DATA VISUALIZATION

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```
df=pd.read_csv("students_placement_data.csv")
df.head()
```

##1-Line Graphs

Draw a line graph between SSC and intermediate/diploma percentage. Does it make sense? Explain

##2-Scatter Plots

Draw a scatter plot between SSC and inter/diploma percentage. Do you find any correlation by just
visual inspection? If yes, then what kind of correlation you can see? (+ve, -ve or No)

```
x=df["SSC Percentage"]
y=df["inter_Diploma_percentage"]
plt.scatter(x,y)
plt.xlabel("SSC scores")
plt.ylabel("inter/diploma scores")
plt.title("scatter plot between ssc and inter/diploma scores")
```

**Here these two variables are possitively correlated**

##3-Histograms

Draw a histogram for B.Tech percentage. How many people have scores between 60-80% in
B.tech. Tell your approximate answer by visual inspection?

```
bins=np.linspace(0,100,11)
print(bins)

plt.hist(df["B.Tech_percentage"],bins,width=9)
plt.xlabel(" B.Tech Score")
```

**Approximately 65 people are having score between 60-80 percentage in btech**

##4-Subplots

Draw a scatter plot between

· intermediate and SSC

· Intermediate and B.Tech

in the same graph. What's your observation?

```
fig,ax=plt.subplots()
ax.scatter(df["inter_Diploma_percentage"],df["SSC Percentage"],label="Inter-ssc")
ax.scatter(df["inter_Diploma_percentage"],df["B.Tech_percentage"],label="Inter-B.Tech")
ax.set_xlabel("Intermediate Score")
ax.legend(loc=2)
```

**people who got good marks in intermediate also got good percentage in ssc 
But,their marks have been reduced in btech as compared to ssc**

##5-Join plots

Draw a jointplot between intermediate and Backlogs in B.tech. What's your observation

```
sns.jointplot(df["inter_Diploma_percentage"],df["Backlogs"],data=df)
```

**1)There are more people with zero backlogs**

**2)people who got good percentage in inter have less backlogs**

**3)Majority people got 75 percentage in intermediate**

##6-Heatmaps

Draw a heatmap between SSC,intermediate and B.tech percentage and find the corelation between them

```
df_sea=df[["inter_Diploma_percentage","SSC Percentage","B.Tech_percentage"]]
df_sea.head()
```

```
sns.heatmap(df_sea.corr(), annot = True)
```

**The correlation between inter and btech is high when compared to ssc**
